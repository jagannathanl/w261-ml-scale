> __b)__ Let $X = (X_0, X_1, ..., X_n)$ be the feature vector. Then the regression equation can be expressed as $\widehat{y} = \theta^T \cdot X $, where $\theta= (\theta_0, \theta_1, ..., \theta_n)$ is the model's parameter vector in the augmented form, containing the bias term $\theta_0$, $\widehat{y}$ is the predicted value of the target and $\theta^T$ is the transpose of $\theta$. Now, the OLS loss $J(\theta)$ function is defined as $J(\theta) = \frac{1}{m}\sum_{i}(\theta^T \cdot X_i - y_i)^2$, where $i$ refers to the $i^{th}$ sample in the training set and $m$ is the number of samples in the training set. In other words, it is mean square distance between the estimated target value and the actual target value $(\widehat{y_i}-y_i)^2$.

> __c)__ In the model above, the feature vector $X$ has $n$ terms with terms $X_0...X_{n-1}$ referring to the $n$ features of the model and the term $X_0$ is always set to one. Most numerical libraries like scikit learn, TensorFlow, pandas etc. are optimized to perform vector and matrix operations effectively. So, it has become a convention in machine learning to represent the data in this format.

> __d)__ In parallel computing, an embarrassingly parallel problem is one where little or no effort is needed to separate the problem into a number of parallel tasks. This is often the case where there is little or no dependency or need for communication between those parallel tasks. As per this definition, OLSLoss is not a embarassingly parallel as mean (average) operation involves sequentially accumulating the sum before we divide it by the count. However, calcuation of individual squared error is an embarassingly parallel.

> __e)__ The variance we computed in part(a) and OLSLoss we computed in part(d) of this problem do match.
```python
# part d - write function to compute loss (FILL IN MISSING CODE BELOW)
def OLSLoss(dataRDD, W):
    """
    Compute mean squared error.
    Args:
        dataRDD - each record is a tuple of (features_array, y)
        W       - (array) model coefficients with bias at index 0
    """
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1]))
    ################## YOUR CODE HERE ##################    
    loss = 0
    loss = augmentedData.map(lambda x: np.dot(W, x[0])-x[1]).map(lambda x: x*x).mean()
    ################## (END) YOUR CODE ##################
    return loss
```

> __a)__  Let $X = (X_0, X_1, ..., X_n)$ be the feature vector. Then the regression equation can be expressed as $\widehat{y} = \theta^T \cdot X $, where $\theta = (\theta_0, \theta_1, ..., \theta_n)$ is the model's parameter vector, containing the bias term $\theta_0$ and $\theta^T$ is the transpose of $\theta$. Now, the gradient is defined as $\nabla(J(\theta)) = \frac{1}{m}\sum_{i} X_i \cdot (\theta^T \cdot X_i- y_i)$, where $i$ refers to the $i^{th}$ sample in the training set and $m$ is the number of samples in the training set. In the gradient equation given above, we can see that the first term inside the summation in the RHS is $X$ and the second term $(y - \widehat{y})$ is the error term. As we divide this sum of this product by $m$ the number of samples, confrms the above Jimi's statement 'the mean of the data weighted by the errors'.

> __b)__ As observed in part(a) of this problem, we compute the gradient by multiplying the error term with the feature vector for each feature vector in the input training dataset. The error term is a scalar that scales the feature vector to yield a partial grandient vector. These partial gradient vectors are later accumulated to form the final gradient. This final accumulation step is a sequential operation.  

> __c)__ _complete the coding portions of this question before answering d & e_ 

> __d)__ Assuming a convex loss function we are guaranteed to find the global minimum if a minimum exists at all. However, this approach is fraught with two problems (1) finding an appropriate learning rate (which is typically a trial and error approach), and (2) components of feature vector components are in different scale. With the second problem, the gradient update will update the gradient quite erratically. As noted in the output of part (c) of this problem, the updated model coefficients swing back and forth. The coefficients of the partial models are swinging between negative and positive values while the maginitude of the coefficients of the model growing exponentially at every step of the iteration. This is indicative of the fact that the components of the feature vector are unnormalized.

> __e)__ Typically, the second problem observed in part(d) could be fixed by normalizing the feature vector. With the normalized feature vector, now the model coefficients are not swinging between positive and negative values and the coefficients are not growing exponentially. Also the good fit of the model for the data is confirmed by the diminishing loss from 0.76 to 0.57.

```python
# part b - function to perform a single GD step
def GDUpdate(dataRDD, W, learningRate = 0.1):
    """
    Perform one OLS gradient descent step/update.
    Args:
        dataRDD - records are tuples of (features_array, y)
        W       - (array) model coefficients with bias at index 0
    Returns:
        new_model - (array) updated coefficients, bias at index 0
    """
    # add a bias 'feature' of 1 at index 0
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1])).cache()
    
    ################## YOUR CODE HERE ################# 
    grad = 2.0*augmentedData.map(lambda x: np.dot(x[0],(np.dot(W, x[0])-x[1]))).mean()
    new_model = W - np.dot(grad, learningRate)
    ################## (END) YOUR CODE ################# 
   
    return new_model
```

> __b)__ We split the dataset into $k$ partitions (or splits), where $k$ is the number of folds. We start with $k$ initial models one model for each partition.
```python
# part b - helper function to emit residuals (RUN THIS CELL AS IS)
def kResiduals(dataPoint, models, splitNum):
    """
    Compute the (squared) residuals for a data point given k different models.
    Note that points from the k-th split are part of the test set for model number k
    and part of the training set for all other models. We'll emit a key to track this.
    Args:
        dataPoint - tuple of (features_array, y)
        models    - list of arrays representing model weights (bias at index 0)
    Returns:
        (stringFormattedKey, squared_error)
    """
    # augment the data point with a bias term at index 0
    X = np.append([1.0], dataPoint[0])
    y = dataPoint[1]
    # emit squared residuals for each model
    for modelNum, W in enumerate(models):
        if modelNum == splitNum:
            yield(f"{modelNum}-test", (W.dot(X) - y)**2)
        else:
            yield(f"{modelNum}-train", (W.dot(X) - y)**2)
```

```python
# part c - fill in the missing code below
def CVLoss(dataSplits, models):
    """
    Compute the k-fold cross-validated test and train loss.
    Args:
        dataSplits - list of RDDs corresponding to the k test splits.
        models     - list of k arrays representing model weights (bias at index 0)
    Returns: 
        tuple of floats: (training_loss, test_loss)
    """
    # compute k residuals for each dataPoint (one for each model)
    partialLossRDD = sc.parallelize([])
    for splitNum, splitRDD in enumerate(dataSplits):
        residuals = splitRDD.flatMap(lambda x: kResiduals(x, models, splitNum))
        partialLossRDD = sc.union([partialLossRDD, residuals])
    
    ################ YOUR CODE HERE #################        
    loss = partialLossRDD.map(lambda x: (x[0], (x[1],1))).reduceByKey(lambda x,y: (x[0]+y[0], x[1]+y[1])).mapValues(lambda x: x[0]/x[1]).collect()
    #print(loss.take(5))
    #[('3-test', 0.7448923280343412), ('1-train', 0.769694341907132), ('0-train', 0.7567388003024996), ('0-test', 0.7744918895571083), ('4-test', 0.7890630594504012)]
    ################ (END) YOUR CODE ################# 
    
    test_loss = np.mean([x[1] for x in loss if x[0].split('-')[1] == 'test'])
    training_loss = np.mean([x[1] for x in loss if x[0].split('-')[1] == 'train'])
    return training_loss, test_loss
```

```python
# part d - helper function RUN THIS CELL AS IS
def partialGradients(splitNum, dataPoint, models):
    """
    Emit partial gradient for this data point for each model.
    NOTE: a data point from split-number k is in the test set for 
    model-k so we don't compute a partial gradient for that model.
    """
    # augment the data point
    X = np.append([1.0], dataPoint[0])
    y = dataPoint[1]
    # emit partial gradients for each model with a counter for averaging later
    for modelNum, W in enumerate(models):
        if modelNum != splitNum:
            yield (modelNum, [(W.dot(X) - y)*X, 1])
```

```python
# part d - perform GD updates for all k models (FILL IN MISSING CODE BELOW)
def CVUpdate(dataSplits, models, learningRate = 0.1):
    """
    Compute gradients for k models given k corresponding dataSplits.
    NOTE: the training set for model-k is all records EXCEPT those in the k-th split.
    """
    # compute partial gradient k-1 times for each fold
    partialsRDD = sc.parallelize([])
    for splitNum, splitRDD in enumerate(dataSplits):
        thisFoldPartialGrads = splitRDD.flatMap(lambda x: partialGradients(splitNum, x, models))
        partialsRDD = sc.union([partialsRDD, thisFoldPartialGrads])

    # compute gradients by taking the average partialGrad for each fold
    gradients = partialsRDD.reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))\
                           .mapValues(lambda x: x[0]/x[1])\
                           .map(lambda x: x[1])\
                           .collect()
    
    # update all k models & return them in a list
    new_models = []
    for W, grad in zip(models, gradients):
        ############# YOUR CODE HERE ############
        new_models.append(W - np.dot(grad, learningRate))
        ############# (END) YOUR CODE ###########
    return new_models
```

```python
# part e - RUN THIS CELL AS IS
def GradientDescent_withCV(dataSplits, wInit, learningRate=0.1, nSteps = 5, verbose = False):
    """
    Train k models in parallel and track cross validated test/train loss.
    Returns:
        train_hist - (list) of floats
        test_hist - (list) of floats
        model_hist - (list) of arrays representing model coefficients (bias at index 0)
    """
    # broadcast initial models (one for each fold)
    bModels = sc.broadcast([wInit] * len(dataSplits))
    
    
    # initialize lists to track performance
    train_loss_0, test_loss_0 = CVLoss(dataSplits, bModels.value)
    train_hist, test_hist, model_hist = [train_loss_0], [test_loss_0], [wInit]
    
    # perform k gradient updates at a time (one for each fold)
    start = time.time()
    for step in range(nSteps):
        new_models = CVUpdate(dataSplits, bModels.value, learningRate)
           
        bModels = sc.broadcast(new_models)

        # log progress
        train_loss, test_loss = CVLoss(dataSplits, bModels.value)
        train_hist.append(train_loss)
        test_hist.append(test_loss)
        model_hist.append(new_models[0])
        
        if verbose:
            print("-------------------")
            print(f"STEP {step}: ")
            print(f"model 1: {[round(w,4) for w in new_models[0]]}")
            print(f" train loss: {round(train_loss,4)}")
            print(f" test loss: {round(test_loss,4)}")
            
    print(f"\n... trained {nSteps} iterations in {time.time() - start} seconds")
    return train_hist, test_hist, model_hist
```

# Question 7: Regularization.

Our goal, as always, is to build a linear model that will extend well to unseen data. Chosing the right combination of features to optimize generalizability can be extremely computationally costly given that there are \\(2^{p}\\) potential models that can be built from \\(p\\) features. Traditional methods like forward selection would involve iteratively testing these options to asses which combinations of features achieve a statistically significant prediction.

Ridge Regression and Lasso Regression are two popular alternatives to OLS, which enable us to train generalizable models without the trouble of forward selection and/or manual feature selection. In practice this 'shrinkage' is achieved by adding a penalty (a.k.a. 'regularization') term to the means squared error loss function. > __a)__ **L2-Regularization (Ridge Regression):** In general, a regularization term is added to penalize magnitude of the coefficient terms. In Ridge regression, a penalty term that is proportional to the sum of squares of $\theta_j$ is used. This regularization form drives the unimportant coefficient to remain small or close to 0. The following two equations shows the loss function and gradient update respectively for the gradient descent:
$$f(\theta, b) = \frac{1}{n} \sum_{i=1}^n [\theta \cdot X_i - y_i]^2 + \lambda \sum_{j=1}^{m}\theta_j^2$$
$$\theta_j = \theta_j - \alpha \cdot (\frac{2}{n} \sum_{i=1}^n [\theta \cdot X_i - y_i] \cdot X_i + 2 \lambda \theta)$$

> __b)__ **L1-Regularization (Lasso Regression):** In Lasso regression, a penalty term that is a sum of absolute values of $\theta_j$ is used. This regularization form tend to completely eliminate the weights of the least important features. In other words Lasso regression automatically performs the feature selection and outputs a sparse model. The following two equations shows the loss and gradient update functions respectively.:
$$f(\theta, b) = \frac{1}{n} \sum_{i=1}^n [\theta \cdot X_i - y_i]^2 + \lambda \sum_{j=1}^{m}|\theta_j|$$
$$\theta_j = \theta_j - \alpha \cdot (\frac{2}{n} \sum_{i=1}^n [\theta \cdot X_i - y_i] \cdot X_i + \lambda(I_{\theta_i>0 \: \forall i>0}(\theta) \cdot 2 - 1))$$  As we noted in the above equation the penalty term in the gradient update essentially assigns a 1 if the weight is greater than 0 and a -1 when the weight is less tha 0.

```python
# part c - gradient descent with regularization
def GDUpdate_wReg(dataRDD, W, learningRate = 0.1, regType = None, regParam = 0.1):
    """
    Perform one gradient descent step/update with ridge or lasso regularization.
    Args:
        dataRDD - tuple of (features_array, y)
        W       - (array) model coefficients with bias at index 0
        learningRate - (float) defaults to 0.1
        regType - (str) 'ridge' or 'lasso', defaults to None
        regParam - (float) regularization term coefficient
    Returns:
        model   - (array) updated coefficients, bias still at index 0
    """
    # augmented data
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1]))
    
    new_model = None
    #################### YOUR CODE HERE ###################
    reg_term = None
    if regType == 'ridge':
        W_ = W.copy()
        W_[0] = 0
        reg_term = regParam * 2 * W_
    elif regType == 'lasso':
        W_ = W.copy()

        W_ = (W_>0).astype(int)*2-1
        W_[0] = 0
        reg_term = regParam * W_
    else:
        reg_term = 0.0
        
    grad = 2.0*augmentedData.map(lambda x: np.dot(x[0],(np.dot(W, x[0])-x[1]))).mean() +reg_term
    new_model = W - np.dot(grad, learningRate)
    ################## (END) YOUR CODE ####################
    return new_model
```

```python
# part c - helper function (RUN THIS CELL AS IS)
def get_residuals(dataRDD, model):
    """
    Return a collected list of tuples (residual, quality_score)
    """
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1]))
    residuals = augmentedData.map(lambda x: (x[1] - model.dot(x[0]), x[1]))
    return residuals.collect()
```

```python
# part d - plotting function (RUN THIS CELL AS IS)
def plotCoeffs(models, featureNames, title):
    """
    Helper Function to show how coefficients change as we train.
    """
    fig, ax = plt.subplots(figsize = (15,8))
    X = list(range(len(models)))
    for data, name in zip(models.T, featureNames):
        if name == "Bias":
            continue
        ax.plot(X, data, label=name)
    ax.plot(X,[0]*len(X), 'k--')
    plt.title(title)
    plt.legend()
    plt.show()
```
